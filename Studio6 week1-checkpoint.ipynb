```python
import numpy as np
import pandas as pd
# Loading data from CSV file
df = pd.read_csv("C:\\Users\\shahe\\Desktop\\Studio 6\\churn group\\customer_churn_telecom_services.csv")
print (df)  
```

```python
# looking at the  missing values are in each column
df.isnull().sum()
```

```python
# Converting to numeric 
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Calculating median
median_value = df['TotalCharges'].median()

# Filling missing values (safely)
df['TotalCharges'] = df['TotalCharges'].fillna(median_value)
```

```python
 #Detecting Outliers (using boxplots)


def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Example: detecting outliers in MonthlyCharges
outliers_monthly = detect_outliers_iqr(df, 'MonthlyCharges')
print(f"Number of outliers in MonthlyCharges: {len(outliers_monthly)}")
```
